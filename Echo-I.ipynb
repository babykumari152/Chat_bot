```python
from nltk.chat.util import Chat, reflections
```

```python
pairs = [
    [
        r"Hi|Hey|Hello",
        ["Hello", "Hey there",]
    ],
    [
        r"How are you ?",
        ["I'm doing good\nHow about You ?",]
    ],
    [
        r"Hi! my name is (.*)",
        ["Hello %1, How are you today ?",]
    ],
     [
        r"What is your name ?",
        ["My name is AIBOT and I'm a chatbot ",]
    ],
    
    [
        r"Sorry (.*)",
        ["Its alright","Its OK, never mind",]
    ],
    [
        r"I'm (.*) doing good",
        ["Nice to hear that","Alright :)",]
    ],
    
    [
        r"(.*) age?",
        ["I'm a computer program dude\nSeriously you are asking me this?",]
        
    ],
    [
        r"what (.*) want ?",
        ["Make me an offer I can't refuse",]
        
    ],
    [
        r"(.*) created ?",
        ["Ms. Baby Kumari created me using Python's NLTK library ","top secret ;)",]
    ],
    [
        r"(.*) (location|city) ?",
        ['Punjab, Chandigarh',]
    ],
    [
        r"how is weather in (.*)?",
        ["Weather in %1 is awesome like always","Too hot man here in %1","Too cold man here in %1","Never even heard about %1"]
    ],
    [
        r"i work in (.*)?",
        ["%1 is an Amazing company, I have heard about it. But they are in huge loss these days.",]
    ],
[
        r"(.*)raining in (.*)",
        ["No rain since last week here in %2","Damn its raining too much here in %2"]
    ],
    [
        r"how (.*) health(.*)",
        ["I'm a computer program, so I'm always healthy ",]
    ],
    [
        r"(.*) (sports|game) ?",
        ["I'm a very big fan of Football",]
    ],
    [
        r"who (.*) sportsperson ?",
        ["Messy","Ronaldo","Roony"]
],
    [
        r"who (.*) (moviestar|actor)?",
        ["Akshay Kumar","Amitav Bachhan","Shahruk Khan"]
],
    [
        r" (.*) ",
        ["I am not getting you dude,let's talk something else!",]
    ],
    
    [
        r"quit",
        ["BBye take care. See you soon :) ","It was nice talking to you. See you soon :)"]
],
]
def welcome():
        print("Hi, I'm AIBOT and I chat alot ;)\n Let's start a conversation. Type quit to leave ") #default message at the start
        chat = Chat(pairs, reflections)
        chat.converse()
if __name__ == "__main__":
    welcome()
```
